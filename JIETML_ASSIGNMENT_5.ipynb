## 2.)DATA PREPARATION

## 2.1) Dropping the column "fnlwgt" !!!

```python
new_df1 = new_df.drop('fnlwgt' , axis =1)
new_df1
```

```python
new_df1.columns
```

```python
new_df1.shape
```

## 2.2) Standardizing the data

```python
new_df1.describe(include = 'object')
```

```python
new_df1.describe(include = 'number')
```

```python
# creating new dataframe of all categorical features

categorical = new_df1.select_dtypes(include=['object'])

categorical.head()
```

```python
# creating the new dataframe of all numerical features

numeric = new_df1.select_dtypes(include=['int64', 'float64'])

numeric.head()
```

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

st_scaler1 = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)

newdf_num = pd.DataFrame(st_scaler1.fit_transform(numeric), 
                      columns = numeric.columns)

newdf_num.head()
```

## 2.3)Encoding the categorical features!!

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

newdf_cat = pd.DataFrame(encoder.fit_transform(categorical), 
                      columns=encoder.get_feature_names(categorical.columns))

newdf_cat.head()
```

```python
# concating numerical and categorical columns

new_df2 = pd.concat([newdf_num, newdf_cat], axis=1)

new_df2.head()
```

## 2.4) Train - test split 

```python
y_data = new_df2.pop('income_>50K')
```

```python
y_data.head()
```

```python
new_df2.head()
```

```python
x = new_df2

y = y_data
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.25, random_state=0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## 3.) Training the model (logistic regression)

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression(max_iter = 300)
classifier.fit(X_train,y_train)
```

## 4.) Predicting the data Model

```python
y_pred = classifier.predict(X_test)
```

```python
pred_prob = classifier.predict_proba(X_test)

np.around(pred_prob, decimals = 2)
```

## 5.)Evaluating the model !!

## 5.1) Using evaluation metrics like:-

### 5.1.1) Accuracy

```python
from sklearn import metrics

metrics.accuracy_score(y_test, y_pred)
```

### OBSERVATIONS :- 
we can observe clearly that accuracy is 84.709 % which is good

### 5.1.2) Confusion metrics

```python
con_metric = metrics.confusion_matrix(y_test, y_pred)
con_metric
```

```python
#ploting heatmap for confusion metric

sns.heatmap(con_metric, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

### Observations
so here as we can observe that :-
<br>total correct predictions made by our model is : 23606 + 5125 = 28731
<br>and total incorrect predictions made by our model : 3283 + 1903 = 5186
<br>so we can conclude that no. of incorrect predictions are quite less than that of correct predicted 
values.

```python
print(metrics.classification_report(y_test,y_pred))

# Number of fetures are 95..!!
```

## 7.) PCA

```python
pca.fit(new_df2)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
Csum = np.cumsum(pca.explained_variance_ratio_)
print(Csum)
```

```python
np.arange(len(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize=(18, 9))
sns.heatmap(X_train.corr(), cmap="YlGnBu")
```

```python
plt.figure(figsize = (12,8))
plt.plot(Csum)
plt.xlabel('number of components')
plt.ylabel('cumulative variance')
plt.show()
```

### observations :
here we can clearly observe that 20 PCA are explaining 90% varicance.

```python
# configuring the parameteres
# number of components = 20
pca_new = PCA(n_components = 20)
pca_new_data = pca_new.fit_transform(new_df2)

# Lets look at the shape of data after PCA
print("shape = ", pca_new_data.shape)
```

```python
pca_df = pd.DataFrame(data=pca_new_data)

pca_df["income"] = y.values

pca_df.head()
```

```python
y_new = pca_df.pop('income')

x_new = pca_df
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_new, y_new, train_size=0.25, random_state=0)
```

```python
# Training the model

from sklearn.linear_model import LogisticRegression 
classifier_pca = LogisticRegression(max_iter = 300)
classifier_pca.fit(X_train_pca,y_train_pca)
```

```python
y_pred_pca = classifier_pca.predict(X_test_pca)
```

```python
y_pred_pca
```

```python
pred_prob_pca = classifier_pca.predict_proba(X_test_pca)

np.around(pred_prob_pca, decimals = 2)
```

```python
#calculating accuracy

from sklearn import metrics

metrics.accuracy_score(y_test_pca, y_pred_pca)
```

```python
# confusion metric

con_metric_pca = metrics.confusion_matrix(y_test_pca, y_pred_pca)
con_metric_pca
```

```python
#ploting heatmap for confusion metric 

sns.heatmap(con_metric_pca, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

### Observations
