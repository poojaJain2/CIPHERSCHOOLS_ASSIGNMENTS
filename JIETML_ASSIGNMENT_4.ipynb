### 7.1) Column Noramlisation using 'MinMaxScaler'

 It is a step of Data Pre Processing which is applied to independent variables or features of data. It basically helps to normalise the data within a particular range .<br>In this approach, the data is scaled to a fixed range - usually 0 to 1.<br>
 A Min-Max scaling is typically done via the following equation:
 <br> Xsc = X - Xmin / Xmax - Xmin <br>
Some examples of algorithms where feature scaling matters are:

i).k-nearest neighbors with an Euclidean distance measure if want all features to contribute equally
<br>ii).k-means (see k-nearest neighbors)
<br>iii).logistic regression, 
 
 
-->For each value in a feature, MinMaxScaler subtracts the minimum value in the feature and then divides by the range. The range is the difference between the original maximum and original minimum. MinMaxScaler preserves the shape of the original distribution.
<br><br> Let's see how is normalisation of columns performed using 'MinMaxScaler'.
