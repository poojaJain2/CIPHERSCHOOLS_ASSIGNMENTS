## 3.) Data Preprocessing 
It includes following tasks :-
<br><br>1.)Remove special characters
<br> 2.)Conversion to lowercase
<br>3.)Tokenization
<br>4.)Stemming or Lemmatization



```python
# import nltk
# nltk.download('stopwords')
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
## initialise the inbuilt Stemmer
stemmer = PorterStemmer()
```

```python
# # We can also use Lemmatizer instead of Stemmer

lemmatizer = WordNetLemmatizer()
```

```python
# initializing the lists

clean_spam_lst = []

spam_len_lst = []
```

```python
def preprocess(raw_spam, flag):
    # Removing special characters and digits
    letters_only = re.sub("[^a-zA-Z]", " ",raw_spam)
    
    # change sentence to lower case
    letters_only = letters_only.lower()

    # tokenize into words
    words = letters_only.split()
    
    # remove stop words                
    words = [w for w in words if not w in stopwords.words("english")]
    
    #Stemming/Lemmatization
    if(flag == 'stem'):
        words = [stemmer.stem(word) for word in words]
    else:
        words = [lemmatizer.lemmatize(word) for word in words]
    
    clean_spam_lst.append(" ".join(words))
    
    spam_len_lst.append(len(words))
```

Stemming makes sure that different variations of a word, say ‘accept’, accepted’, ‘accepting’ and ‘acceptance,’ are represented by a single token - ‘accept’, because they all represent the same information (represented by the 'stem' of the word).

```python
from tqdm import tqdm, tqdm_notebook

tqdm.pandas()
```

```python
#Stemmatizing the data

clean_spam_lst = []

spam_len_lst = []

df['message'].progress_apply(lambda x: preprocess(x, 'stem'))

df['clean_spam_stem'] = clean_spam_lst

df['spam_length_stem'] = spam_len_lst
```

```python
# Lemmatizing the data

clean_spam_lst = []

spam_len_lst = []

df['message'].progress_apply(lambda x: preprocess(x, 'lemma'))

df['clean_spam_lema'] = clean_spam_lst

df['spam_length_lema'] = spam_len_lst
```

## 4.1) Train - Test split 

```python

from sklearn.model_selection  import train_test_split

train, test = train_test_split(df,test_size=0.2,random_state=42)
```

```python
train_clean_spam=[]

for msg in train['clean_spam_stem']:
    train_clean_spam.append(msg)

test_clean_spam=[]

for msg in test['clean_spam_stem']:
    test_clean_spam.append(msg)
```

## 4.2) Identifying the target variable

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word")

train_features = vectorizer.fit_transform(train_clean_spam)

test_features = vectorizer.transform(test_clean_spam)
```

## 6). Evaluation of the model
<b>For each Model, evaluate using following metrics:</b>
<b><br>i).Check accuracy score
<br>ii).Building confusion metric

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# These are all the classes of the algorithms which we're gonna use here ..!!
# i.e from library 'sklearn' we are importing module named - 'linear_model' for 'logisticRegression'
# From library 'sklearn' we are importing module named - 'tree' for ' DecisionTreeClassifier'
# From library 'sklearn' we are importing module named - 'svm' for 'SVC'..!!!!
```

```python
# for obtaining "Accuracy" we gonna import 'accuracy_score' class from 'metrics' Module..!!

from sklearn.metrics import accuracy_score

# for obtaining "confusion_matrix "  we gonna import 'confusion_matrix ' class from 'metrics' Module..!!
from sklearn.metrics import confusion_matrix 
 
# for obtaining "classification_report"  we gonna import 'classification_report ' class from 'metrics' Module..!!
from sklearn.metrics import classification_report 
```

```python
Classifiers = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    SVC()]
```

```python
# Converting sparse matrix into dense matrix..!!

dense_features = train_features.toarray()

dense_test = test_features.toarray()
```

```python
for classific in tqdm(Classifiers):
    fit = classific.fit(dense_features,train['target'])
    
    # prediciting the model
    
    pred = fit.predict(dense_test)
    
    # calculate accuracy ..!!
    accuracy = accuracy_score(pred,test['target'])
    print('o'*35+classific.__class__.__name__+'o'*35) 
    print("\n 6.1) CHECK ACCURACY SCORE.......!!!")
    print('Accuracy of '+classific.__class__.__name__+' is '+str(accuracy)) 
    
    # obtaining confusion metrics..!!
    print("\n 6.2.1).CONFUSION METRIC......!!!")
    con_metric = confusion_matrix(test['target'], pred)
    print(con_metric)
    
    #ploting heatmap for confusion metric 
    print("\n 6.2.2). PLOTTING HEATMAP.....!!! ")
    sns.heatmap(con_metric, annot=True, fmt='d')
    plt.title("Confusion Matrix")
    plt.show()
    
    # printing classification report..!!
    print("\n 6.3) PRINTING CLASSIFICATION_REPORT.......!!!")    
    print(classification_report(test['target'], pred))
```
